## Regression with Keras


```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

<a id='item33'></a>


## Build a Neural Network


Let's define a function that defines our regression model for us so that we can conveniently call it to create our model.


```python
# define regression model, generic function
def regression_model(num_hidden_layer,num_node,n_cols):
    # create model
    model = Sequential()
    
    # create hidden layers
    for index in range(num_hidden_layer):
        if index==0:
            model.add(Dense(num_node[index], activation='relu', input_shape=(n_cols,)))
        else:
            model.add(Dense(num_node[index], activation='relu'))
            
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

The above function is a generic function to create a model that has N hidden layers, each of N hidden units.

inputs : num_hidden_layer,num_node,n_cols

## A. Build a baseline model (5 marks) 
Use the Keras library to build a neural network with the following:

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.

1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn.

2. Train the model on the training data using 50 epochs.

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

5. Report the mean and the standard deviation of the mean squared errors.


```python
#import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

print("predictors ", predictors.shape)
print("target ", target.shape)

# splitting predictors,target  // 30% of the data for testing
predictors_train,predictors_test,target_train,target_test = train_test_split(predictors,target,test_size = 0.3,random_state=4)

print("predictors_train ", predictors_train.shape)
print("predictors_test ", predictors_test.shape)

# number of predictors_train_cols
predictors_train_cols = predictors_train.shape[1]

# number of predictors_test_cols
predictors_test_cols = predictors_test.shape[1]
```

```python
# list of 50 mean squared errors. repeat 1-3
N=50
MS_Error  = np.zeros((N))
for ind in range(N):
    predictors_train,predictors_test,target_train,target_test = train_test_split(predictors,target,test_size = 0.3,random_state=4)
    # One hidden layer of 10 nodes, and a ReLU activation function
    # build the model
    model = regression_model(1,[10],predictors_train_cols)
    # Train the model on the training data using 50 epochs.
    model.fit(predictors_train, target_train, validation_split=0.3, epochs=50, verbose=2)
    # Evaluate the model on the test data 
    predicted_target_values = model.predict(predictors_test)
    # compute the mean squared error between the predicted concrete 
    # strength and the actual concrete strength
    # predicted concrete strength  : predicted_target_values
    # actual concrete strength  :  target_test
    MS_Error[ind] = mean_squared_error(target_test, predicted_target_values)
```

```python
# Report the mean and the standard deviation of the mean squared errors.
print("Mean of MSE : ",np.mean(MS_Error))
print("SD of MSE :", np.std(MS_Error))
```
